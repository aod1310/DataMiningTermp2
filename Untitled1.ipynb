```python
import pandas as pd
import numpy as np
import os
import pickle
from apyori import apriori
import time
```

```python
path = './ml-25m/'

movies = pd.read_csv(os.path.join(path, 'movies.csv'))
ratings = pd.read_csv(os.path.join(path, 'ratings.csv'))
movie_rating = pd.merge(ratings, movies)

# data cleaning
movie_rating = movie_rating.drop('timestamp', axis=1)
movie_rating = movie_rating.drop(movie_rating.loc[movie_rating['genres']=='(no genres listed)'].index, axis=0)

mania_users = None

with open('like_provocative45.users', 'rb') as f:
    mania_users = pickle.load(f)

print('mania users : {}'.format(len(mania_users)))
    
mania = pd.DataFrame([])
for i, user in enumerate(mania_users):
    mania = mania.append(movie_rating[movie_rating['userId']==user])
mania
```

```python
mania_copy = mania.copy()
mania_copy = mania_copy.reset_index()
s = time.time()
for i in range(len(mania_copy['genres'])):
    splited = mania_copy['genres'][i].split('|')
    mania_copy['genres'][i] = splited
    if i % 10000 == 0:
        print(i, splited)
print(time.time()-s)
```

```python
mania_copy
```

```python
genres = mania_copy['genres']
genres
```

```python
s= time.time()
association_rules = apriori(genres, min_support=0.05, min_confidence=0.6, min_left=3, min_length=2)
association_rlt = list(association_rules)
print(time.time()-s)
```

```python
len(association_rlt)
```

```python
for item in association_rlt:
    # first index of the inner list
    # Contains base item and add item
    pair = item[0]
    items = [x for x in pair]
    print(len(items))
    print("Rule: " + items[0] + ' genre is with ' + str(items[1:]))

    # second index of the inner list
    print("Support: " + str(item[1]))

    # third index of the list located at 0th
    # of the third index of the inner list

    print("Confidence: " + str(item[2][0][2]))
    print("Lift: " + str(item[2][0][3]))
    print("=====================================")
```

```python
supports = [item[1] for item in association_rlt]
max_sup = supports.index(max(supports))
supports.pop(3)
supports.pop(4)
supports
```

```python
confidences = [item[2][0][2] for item in association_rlt]
max_conf = confidences.index(max(confidences))
confidences.pop(3)
confidences.pop(4)
confidences
```

```python
genre_rlt = [list(item[0]) for item in association_rlt]
genre_rlt
```

```python
# remove children movie
for i, rlt in enumerate(genre_rlt):
    if 'Children' in rlt:
        print(rlt)
        genre_rlt.remove(genre_rlt[i])
genre_rlt
```

```python
rate4genre = [ [] for _ in range(len(genre_rlt))]
for i, g in enumerate(genre_rlt):
    g = set(g)
    print(i, g)
    for j, row in enumerate(mania_copy.iterrows()):
        l = set(row[1]['genres'])
        if g.issubset(l):
            rate4genre[i].append(float(row[1]['rating']))
    rate4genre[i] = np.mean(rate4genre[i])
```

```python
rate4genre
```

```python
max_conf_idx = confidences.index(max(confidences))
max_sup_idx = supports.index(max(supports))

for idx in range(len(genre_rlt)):
    rlt = supports[idx] * confidences[idx] * rate4genre[idx]
    print(genre_rlt[idx], 'success score : ', rlt)
print('best support was {}, {}'.format(genre_rlt[max_sup_idx], max(supports)))
print('best confience was {}, {}'.format(genre_rlt[max_conf_idx], max(confidences)))
```

```python
best_scores = [['Action', 'Sci-Fi'], ['Adventure', 'Action', 'Sci-Fi']]
titles = [[], []]
for i, g in enumerate(best_scores):
    g = set(g)
    print(i, g)
    for j, row in enumerate(mania_copy.iterrows()):
        l = set(row[1]['genres'])
        if g.issubset(l):
            titles[i].append(row[1]['title'])
```
